Getting Images and Labels for each item in the dataset

```python
def get_label(file_path):
    return tf.strings.split(file_path, os.path.sep)[-2]
```

```python
def process_image(file_path):
    label = get_label(file_path)
    
    img = tf.io.read_file(file_path)
    img=tf.image.decode_jpeg(img)
    img=tf.image.resize(img, [360,360])

    return img, label
```

```python
for t in train_ds.take(4):
    print(t.numpy())
```

```python
train_ds = train_ds.map(process_image)
for img, label in train_ds.take(3):
    print("Image: ", img)
    print("Label: ",label)
```

```python
test_ds = test_ds.map(process_image)
for img, label in test_ds.take(3):
    print("Image: ", img)
    print("Label: ",label)
```

```python
val_ds = val_ds.map(process_image)
for img, label in val_ds.take(3):
    print("Image: ", img)
    print("Label: ",label)
```

Scale function

```python
def scale(image, label):
    return image/255, label
```

```python
train_ds = train_ds.map(scale)
for image, label in train_ds.take(5):
   print ("****Image: ",image.numpy()[0][0])
   print ("****Label: ",label.numpy() )
```
