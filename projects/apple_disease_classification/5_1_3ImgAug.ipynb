```python
import imgaug.augmenters as iaa
import cv2
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

LOAD IMAGES

```python
blotch_train_images = []
blotch_train_files = glob.glob("data/Train/Blotch_Apple/*.jpg")

for img_path in blotch_train_files:
    img = cv2.imread(img_path)
    blotch_train_images.append(img) # append each image to the blotch train images
# print (blotch_train_images)
```

```python
print (len(blotch_train_images))
```

AUGMENTATION

```python
augmentation = iaa.Sequential([
     # 1. Flip
    iaa.Fliplr(0.5),
    iaa.Flipud(0.5),
    # 2. Affine
    iaa.Affine(translate_percent={"x": (-0.1, 0.1), "y": (-0.1, 0.1)},
               rotate=(-20, 20),
               scale=(0.5, 1.5)), #helps the model to learn different sizes of the apples
    # 3. Multiply - both multiply and linearcontrast are to adapt the colors of the apples
    iaa.Multiply((0.8, 1.2)),
    # 4. Linearcontrast
    iaa.LinearContrast((0.6, 1.4)),
    
], random_order=True)

augmented_blotch_train_images = augmentation(images=blotch_train_images)
```

SHOW IMAGES
